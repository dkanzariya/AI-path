```
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage

with open('plot_dnn_8x8.npy', 'rb') as f:
    error1_88 = np.load(f)
    error2_88 = np.load(f)
    SNR = np.load(f)

with open('plot_dnn_2x2.npy', 'rb') as f:
    error1_22 = np.load(f)
    error2_22 = np.load(f)
    SNR = np.load(f)

with open('plot_dnn_4x4.npy', 'rb') as f:
    error1_44 = np.load(f)
    error2_44 = np.load(f)
    SNR = np.load(f)

error44 = scipy.ndimage.gaussian_filter(error1_88, 19)
error88 = scipy.ndimage.gaussian_filter(error1_44, 19)
error22 = scipy.ndimage.gaussian_filter(error1_22, 19)
error244 = scipy.ndimage.gaussian_filter(error2_88, 19)
error288 = scipy.ndimage.gaussian_filter(error2_44, 19)
error222 = scipy.ndimage.gaussian_filter(error2_22, 19)

name = "SER_MIMO(NXN)_NOMA__DNN_CNN_BPSK"
fig, ax = plt.subplots()
ax.axis([ 1,25, 10**-6,1])
ax.semilogy(SNR, error44, label='user_good_channel_dnn_44')
ax.semilogy(SNR, error88, label='user_bad_channel_dnn_88')
ax.semilogy(SNR, error22, label='user_bad_channel_dnn_22')
ax.semilogy(SNR, error244, label='user_good_channel_dnn_244')
ax.semilogy(SNR, error288, label='user_bad_channel_dnn288')
ax.semilogy(SNR, error222, label='user_bad_channel_cnn222')

plt.legend()
plt.grid()
plt.xlabel('SNR')
plt.ylabel('SER')
plt.title(name)
plt.savefig(name)
plt.savefig(name, format='eps')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage

with open('/content/plot_accuracy_cnn_8x8.npy', 'rb') as f:
    acc1 = np.load(f)
    acc2 = np.load(f)

with open('/content/plot_loss_cnn_8x8.npy', 'rb') as f:
    loss1 = np.load(f)
    loss2 = np.load(f)

name = "NOMA_MIMO(8x8)_accuracy"
fig, ax = plt.subplots()
# ax.axis([ 1,21, 10**-8,1])
acc1 = scipy.ndimage.gaussian_filter(acc1, 10)
acc2 = scipy.ndimage.gaussian_filter(acc2, 20)
plt.plot(acc1, label = "user_good_channel")
plt.plot(acc2, label = "user bad_channel")
name = "NOMA_MIMO(8x8)_loss"
# fig, ax = plt.subplots()
# ax.axis([ 1,21, 10**-8,1])
loss1 = scipy.ndimage.gaussian_filter(loss1, 5)
loss2 = scipy.ndimage.gaussian_filter(loss2, 10)
plt.plot(loss1, label = "user_good_channel")
plt.plot(loss2, label = "user bad_channel")
plt.legend()
# plt.grid()
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title(name)
plt.savefig(name)
plt.savefig(name, format='eps')
plt.show()
```
